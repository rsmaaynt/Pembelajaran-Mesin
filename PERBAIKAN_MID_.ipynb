```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/Dataset UTS kelas A.csv')
data
```

```python
data1 = data[['year','selling_price']]
data1
```

```python
data1.info()
```

```python
data1.describe()
```

```python
data1.corr()
```

```python
plt.scatter(data1['year'],data1['selling_price'])
plt.xlabel('year')
plt.ylabel('selling_price')
plt.show()
```

```python
x = data1.iloc[:, [0]].values
y = data1.iloc[:, [1]].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Model Regresi (Training)

```python
model = LinearRegression()
model.fit (x_train, y_train)
```

```python
print(f'Coefficients: {model.coef_}')
print(f'Intercept: {model.intercept_}')
```

Y = -1.11036073e+08 + 55405.24907089

```python
# sebaran data train
plt.scatter(x_train, y_train)
plt.plot(x_train, model.predict(x_train), c='red')
plt.xlabel('year')
plt.ylabel('selling_price')
plt.show()
```

Testing model

```python
y_prediksi = model.predict(x_test)
print(y_prediksi)
```

Evaluasi Model

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print('MSE:', mean_squared_error(y_test, y_prediksi))
print('R2:' , r2_score(y_test, y_prediksi))
print('MAE:', mean_absolute_error(y_test, y_prediksi))
```

Test Data Baru

```python
data_baru =[[2019]]
prediksi_baru = model.predict(data_baru)
print(prediksi_baru)
```

Visualisasi Data

```python
# Visualisasi hasil prediksi data training
plt.figure(figsize=(10,5))
plt.title('visualisasi hasil prediksi data training')

# data observasi (data train) warna biru
plt.scatter(x_train,y_train,c='blue')

# data hasil prediksi dari data train
plt.plot(x_train, model.predict(x_train),c='red')
plt.xlabel('year')
plt.ylabel('selling_price')
plt.show()
```
