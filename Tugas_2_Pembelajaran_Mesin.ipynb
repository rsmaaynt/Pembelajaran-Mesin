PIE CHART

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = '/Quality of life index by countries 2020.csv'
data = pd.read_csv(dataset)
data
```

```python
data_top = data.head(5)
data_top
```

```python
labels = data_top['Country']
sizes = data_top['Quality of Life Index']
plt.figure(figsize=(7, 5))  # Ukuran figure, bisa disesuaikan
explode = (0.1, 0, 0, 0, 0)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140,explode=explode,shadow=True)
plt.axis('equal')  # Pastikan bahwa pie di-draw sebagai lingkaran.
plt.title('Top 5 Quality of Life Index by Country')  # Judul pie chart

# Menampilkan pie chart
plt.show()
```
